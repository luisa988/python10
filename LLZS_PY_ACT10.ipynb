 REALIZA UN PROGRAMA QUE UTILICE UNA LISTA DE DICCIONARIOS



MENU

1.- Agregar (automatico)

2.- Agregar (manual)

3.- Imprimir lista

4.- Buscar {ID}

5.- Ordenar

6.- Eliminar {ID}

7.- Borrar Toda la lista {ID}

0.- SALIR



NOTA: Los datos del diccionario son los Datos básicos de un trabajador de una fabrica.

```python
lista_trabajadores = []

def agregar_automatico():
    datos_trabajador = {
        "ID": 1,
        "Nombre": "Juan",
        "Apellido": "Pérez",
        "Cargo": "Operario"
    }
    lista_trabajadores.append(datos_trabajador)
    print("Trabajador agregado automáticamente.")

def agregar_manual():
    id_trabajador = int(input("Ingrese el ID del trabajador: "))
    nombre = input("Ingrese el nombre del trabajador: ")
    apellido = input("Ingrese el apellido del trabajador: ")
    cargo = input("Ingrese el cargo del trabajador: ")
    datos_trabajador = {
        "ID": id_trabajador,
        "Nombre": nombre,
        "Apellido": apellido,
        "Cargo": cargo
    }
    lista_trabajadores.append(datos_trabajador)
    print("Trabajador agregado manualmente.")

def imprimir_lista():
    if len(lista_trabajadores) == 0:
        print("La lista de trabajadores está vacía.")
    else:
        print("Lista de trabajadores:")
        for trabajador in lista_trabajadores:
            print(f"ID: {trabajador['ID']}, Nombre: {trabajador['Nombre']}, Apellido: {trabajador['Apellido']}, Cargo: {trabajador['Cargo']}")

def buscar_por_id():
    id_buscar = int(input("Ingrese el ID del trabajador a buscar: "))
    encontrado = False
    for trabajador in lista_trabajadores:
        if trabajador["ID"] == id_buscar:
            print("Trabajador encontrado:")
            print(f"ID: {trabajador['ID']}, Nombre: {trabajador['Nombre']}, Apellido: {trabajador['Apellido']}, Cargo: {trabajador['Cargo']}")
            encontrado = True
            break
    if not encontrado:
        print("No se encontró ningún trabajador con ese ID.")

def ordenar_por_id():
    lista_trabajadores.sort(key=lambda x: x["ID"])
    print("La lista de trabajadores ha sido ordenada por ID.")

def eliminar_por_id():
    id_eliminar = int(input("Ingrese el ID del trabajador a eliminar: "))
    eliminado = False
    for trabajador in lista_trabajadores:
        if trabajador["ID"] == id_eliminar:
            lista_trabajadores.remove(trabajador)
            print("Trabajador eliminado.")
            eliminado = True
            break
    if not eliminado:
        print("No se encontró ningún trabajador con ese ID.")

def borrar_lista():
    lista_trabajadores.clear()
    print("La lista de trabajadores ha sido borrada.")

while True:
    print("MENU\n")
    print("1.- Agregar (automático)")
    print("2.- Agregar (manual)")
    print("3.- Imprimir lista")
    print("4.- Buscar por ID")
    print("5.- Ordenar")
    print("6.- Eliminar por ID")
    print("7.- Borrar toda la lista")
    print("0.- SALIR\n")

    opcion = int(input("Ingrese una opción: "))

    if opcion == 1:
        agregar_automatico()
    elif opcion == 2:
        agregar_manual()
    elif opcion == 3:
        imprimir_lista()
    elif opcion == 4:
        buscar_por_id()
    elif opcion == 5:
        ordenar_por_id()
    elif opcion == 6:
        eliminar_por_id()
    elif opcion == 7:
        borrar_lista()
    elif opcion == 0:
        break
    else:
        print("Opción inválida. Intente nuevamente.")

print("Programa finalizado.")
```
